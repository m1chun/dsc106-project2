```python
import altair as alt

# Convert % columns to numeric if they still contain strings
for col in ['women_no_clinic_pct', 'out_of_state_pct']:
    abortion[col] = (
        abortion[col]
        .astype(str)
        .str.replace('%', '', regex=False)
        .astype(float)
    )

# Create scatter plot
chart = (
    alt.Chart(abortion)
    .mark_circle(size=125)
    .encode(
        x=alt.X('women_no_clinic_pct:Q',
                title='% Women Aged 15-44 Living In A County Without A Clinic (2020)', 
                axis=alt.Axis(titleFontSize=14, titlePadding=12)),
        y=alt.Y('out_of_state_pct:Q',
                title='% Residents Obtaining Abortions Out-of-State (2020)',
                axis=alt.Axis(titleFontSize=14, titlePadding=12)),
        color=alt.Color('election_2024:N',
                        scale=alt.Scale(
                            domain=['Democrat', 'Republican'],
                            range=['#3b82f6', '#ef4444']
                        ),
                        title='2024 Election Result'),
        tooltip=[alt.Tooltip('state', title = 'State'), alt.Tooltip('women_no_clinic_pct', title = '% Women Without Clinic Access'), 
                 alt.Tooltip('out_of_state_pct', title='% Abortions Received Out-of-State'), alt.Tooltip('election_2024', title = '2024 Presidential Election Result')]
    )
    .properties(
        title='Women living in Republican states are travelling at high rates to receive abortions',
        width=600,
        height=400
    )
    .configure_legend(
    titleFontSize=14,
    labelFontSize=13,
    symbolSize=200
    )
    .configure_axis(labelFontSize=12, titleFontSize=14)
    .configure_title(fontSize=20, anchor='start', offset=14)
).interactive()

chart.save('final_viz1.json')
```

```python
import altair as alt

# Scatter plot: Clinic change vs. Abortion rate change by state
scatter_chart = (
    alt.Chart(abortion)
    .mark_circle(size=125)  # match point size
    .encode(
        x=alt.X(
            'clinic_change_pct:Q',
            title='% Change in Number of Abortion Clinics (2017-2020)',
            axis=alt.Axis(titleFontSize=14, titlePadding=12)
        ),
        y=alt.Y(
            'abortion_rate_change_pct:Q',
            title='% Change in Number of Abortions (2017-2020)',
            axis=alt.Axis(titleFontSize=14, titlePadding=12)
        ),
        color=alt.Color(
            'election_2024:N',
            scale=alt.Scale(
                domain=['Democrat', 'Republican'],
                range=['#3b82f6', '#ef4444']
            ),
            title='2024 Election Result',
            legend=alt.Legend(
                titleFontSize=14,
                labelFontSize=13,
                symbolSize=200
            )
        ),
        tooltip=[
            alt.Tooltip('state', title='State'),
            alt.Tooltip('clinic_change_pct', title='Change in Clinics (%)'),
            alt.Tooltip('abortion_rate_change_pct', title='Change in Abortion Rate (%)'),
            alt.Tooltip('election_2024', title='2024 Presidential Election Result')
        ]
    )
    .properties(
        width=600,
        height=400,
        title=alt.TitleParams(
            ["The political war over abortion access seems pointless—",
             "changes in the number of abortions and clinics are uniform across red and blue states"])
    )
    .configure_legend(
        titleFontSize=14,
        labelFontSize=13,
        symbolSize=200
    )
    .configure_axis(
        labelFontSize=12,
        titleFontSize=14
    )
    .configure_title(
        fontSize=20,
        anchor='start',
        offset=14
    )
).interactive()

scatter_chart.save('final_viz2.json')
```

```python
import pandas as pd
import altair as alt

# Example reshape
abortion_melted = abortion.melt(
    id_vars=['state', 'election_2024'],
    value_vars=['clinic_change_pct', 'abortion_rate_change_pct'],
    var_name='Metric',
    value_name='Percent Change'
)

# Create a grouped bar chart
bar_chart = (
    alt.Chart(abortion_melted)
    .mark_bar()
    .encode(
        x=alt.X('Metric:N', title='Metric'),
        y=alt.Y('Percent Change:Q', title='Percent Change (2017–2020)'),
        color=alt.Color(
            'election_2024:N',
            scale=alt.Scale(
                domain=['Democrat', 'Republican'],
                range=['#3b82f6', '#ef4444']
            ),
            title='2024 Election Result'
        ),
        column=alt.Column('election_2024:N', title='2024 Election Result'),
        tooltip=['state', 'Metric', 'Percent Change']
    )
    .properties(
        width=250,
        height=300,
        title='Changes in Clinic Access and Abortion Rates (2017–2020)'
    )
    .configure_axis(labelFontSize=12, titleFontSize=14)
    .configure_title(fontSize=18, anchor='start')
)

bar_chart
```

```python
import altair as alt
import pandas as pd
from vega_datasets import data

# Load US states topojson
states = alt.topo_feature(data.us_10m.url, 'states')

# Example: your abortion DataFrame already has these columns
# abortion.columns → ['U.S. State', 'No. of abortions per 1,000 women aged 15–44, by state of residence, 2020',
#                     '% of counties without a known abortion provider, 2014', 'Election2024']

# (optional) simplify column names for easier use
abortion = abortion.rename(columns={
    'U.S. State': 'State',
    'No. of abortions per 1,000 women aged 15–44, by state of residence, 2020': 'Abortions_per_1000',
    '% of counties without a known abortion provider, 2014': 'Percent_no_provider'
})

# --- Add FIPS codes (needed to link to the topojson) ---
# Mapping of state names to FIPS codes (50 states + DC)
state_fips = {
    "Alabama": 1, "Alaska": 2, "Arizona": 4, "Arkansas": 5, "California": 6,
    "Colorado": 8, "Connecticut": 9, "Delaware": 10, "District of Columbia": 11,
    "Florida": 12, "Georgia": 13, "Hawaii": 15, "Idaho": 16, "Illinois": 17,
    "Indiana": 18, "Iowa": 19, "Kansas": 20, "Kentucky": 21, "Louisiana": 22,
    "Maine": 23, "Maryland": 24, "Massachusetts": 25, "Michigan": 26,
    "Minnesota": 27, "Mississippi": 28, "Missouri": 29, "Montana": 30,
    "Nebraska": 31, "Nevada": 32, "New Hampshire": 33, "New Jersey": 34,
    "New Mexico": 35, "New York": 36, "North Carolina": 37, "North Dakota": 38,
    "Ohio": 39, "Oklahoma": 40, "Oregon": 41, "Pennsylvania": 42,
    "Rhode Island": 44, "South Carolina": 45, "South Dakota": 46, "Tennessee": 47,
    "Texas": 48, "Utah": 49, "Vermont": 50, "Virginia": 51, "Washington": 53,
    "West Virginia": 54, "Wisconsin": 55, "Wyoming": 56
}

abortion['id'] = abortion['State'].map(state_fips)

# --- Create choropleth map ---
chart = (
    alt.Chart(states)
    .mark_geoshape(stroke='white')
    .encode(
        color=alt.Color(
            'Abortions_per_1000:Q',
            scale=alt.Scale(scheme='blues'),
            title='Abortions per 1,000 Women (2020)'
        ),
        tooltip=[
            'State:N',
            'Abortions_per_1000:Q',
            'Percent_no_provider:Q',
            'Election2024:N'
        ]
    )
    .transform_lookup(
        lookup='id',
        from_=alt.LookupData(
            abortion,
            key='id',
            fields=['State', 'Abortions_per_1000', 'Percent_no_provider', 'Election2024']
        )
    )
    .project('albersUsa')
    .properties(
        width=700,
        height=450,
        title='Abortion Rate per 1,000 Women (2020) by State'
    )
)
chart
```
