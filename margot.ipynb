```python
data = pd.read_excel("Abortion-Data.xlsx")
```

```python
state_results_2024 = {
    "Alabama": "Republican",
    "Alaska": "Republican",
    "Arizona": "Republican",
    "Arkansas": "Republican",
    "California": "Democrat",
    "Colorado": "Democrat",
    "Connecticut": "Democrat",
    "Delaware": "Democrat",
    "District of Columbia": "Democrat",
    "Florida": "Republican",
    "Georgia": "Republican",
    "Hawaii": "Democrat",
    "Idaho": "Republican",
    "Illinois": "Democrat",
    "Indiana": "Republican",
    "Iowa": "Republican",
    "Kansas": "Republican",
    "Kentucky": "Republican",
    "Louisiana": "Republican",
    "Maine": "Democrat",
    "Maryland": "Democrat",
    "Massachusetts": "Democrat",
    "Michigan": "Republican",
    "Minnesota": "Democrat",
    "Mississippi": "Republican",
    "Missouri": "Republican",
    "Montana": "Republican",
    "Nebraska": "Republican",
    "Nevada": "Republican",
    "New Hampshire": "Democrat",
    "New Jersey": "Democrat",
    "New Mexico": "Democrat",
    "New York": "Democrat",
    "North Carolina": "Republican",
    "North Dakota": "Republican",
    "Ohio": "Republican",
    "Oklahoma": "Republican",
    "Oregon": "Democrat",
    "Pennsylvania": "Republican",
    "Rhode Island": "Democrat",
    "South Carolina": "Republican",
    "South Dakota": "Republican",
    "Tennessee": "Republican",
    "Texas": "Republican",
    "Utah": "Republican",
    "Vermont": "Democrat",
    "Virginia": "Democrat",
    "Washington": "Democrat",
    "West Virginia": "Republican",
    "Wisconsin": "Republican",
    "Wyoming": "Republican",
}
```

```python
data['Election2024'] = data['U.S. State'].map(state_results_2024)

data = data.rename(columns={
    'U.S. State': 'state',
    '% change in abortion rate, 2017-2020': 'change_rate_17_20',
    '% change in the no. of abortion clinics, 2017-2020': 'change_clinics_17_20',
    '% change in the no. of abortion providers, 2014-2017': 'change_providers_14_17',
    '% of all U.S. abortions, by state of occurrence, 2020': 'percent_us_abortions_2020',
    '% of counties without a known abortion provider, 2014': 'counties_no_provider_2014',
    '% of counties without a known clinic, 2020': 'counties_no_clinic_2020',
    '% of residents obtaining abortions who traveled out of state for care, 2020': 'out_of_state_abortions_2020',
    '% of women aged 15-44 living in a county without a clinic, 2020': 'women_no_clinic_2020',
    '% of women aged 15-44 living in a county without an abortion provider, 2014': 'women_no_provider_2014',
    'No. of abortions per 1,000 women aged 15–44, by state of residence, 2020': 'abortions_per_1000_women_2020',
    'No. of abortions, by state of occurrence, 2020': 'abortions_by_occurrence_2020',
    'No. of abortions, by state of residence, 2020': 'abortions_by_residence_2020',
    'No. of federally funded abortions, 2010': 'federal_abortions_2010',
    'No. of state funded abortions, 2010': 'state_abortions_2010',
    'Reported public expenditures for abortions (in 000s of dollars), federal, 2015': 'federal_expenditures_2015',
    'Reported public expenditures for abortions (in 000s of dollars), state, 2015': 'state_expenditures_2015',
    'Total no. of publicly funded abortions , 2010': 'public_abortions_total_2010',
    'Total reported public expenditures for abortions (in 000s of dollars), 2015': 'public_expenditures_total_2015',
    'Election2024': 'election_2024'
})
```

```python
chart = (
    alt.Chart(data)
    .mark_circle(size=100)
    .encode(
        x=alt.X('state_abortions_2010', title='State abortions'),
        y=alt.Y('out_of_state_abortions_2020', title='Percent of Out-Of-State Abortions in 2020'),
        color=alt.Color('election_2024', scale=alt.Scale(domain=['Democrat', 'Republican'], range=['blue', 'red']), title='State Lean'),
        tooltip=['state', 'election_2024']
    )
    .properties(
        title='Clinic-less Counties lead to more Out-of-State',
        width=600,
        height=400
    )
)

chart
```

```python
chart = (
    alt.Chart(data)
    .mark_circle(size=100)
    .encode(
        x=alt.X('abortions_by_residence_2020', title='Abortions by residence 2020'),
        y=alt.Y('abortions_by_occurrence_2020', title='State Abortions'),
        color=alt.Color('election_2024', scale=alt.Scale(domain=['Democrat', 'Republican'], range=['blue', 'red']), title='State Lean'),
        tooltip=['state', 'election_2024']
    )
    .properties(
        title='',
        width=600,
        height=400
    )
)

chart
```

```python
data_long = data.melt(id_vars=['state', 'election_2024'],
                  value_vars=['women_no_clinic_2020', 'abortions_per_1000_women_2020'],
                  var_name='metric', value_name='value')

metric_labels = {
    'women_no_clinic_2020': 'No access to clinic',
    'abortions_per_1000_women_2020': 'Abortions per 1,000 women'
}
data_long['metric'] = data_long['metric'].map(metric_labels)


chart_slope = (
    alt.Chart(data_long)
    .mark_line(point=True, strokeWidth=2)
    .encode(
        x=alt.X('metric:N', title='Metric'),
        y=alt.Y('value:Q', title='Value'),
        color=alt.Color('election_2024:N',
                        scale=alt.Scale(domain=['Democrat', 'Republican'], range=['blue', 'red']),
                        title='State Lean'),
        detail='state:N',
        tooltip=['state', 'election_2024', 'metric', 'value']
    )
    .properties(
        title='“Access vs. Outcome” — The Abortion Access Paradox',
        width=500,
        height=400
    )
)

chart_slope
```
